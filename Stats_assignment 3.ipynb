Q1. Blood glucose levels for obese patients have a mean of 100 with a standard deviation of 15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an effect or not.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as st
from scipy.stats import norm
import math
```

```python
def hypothesis_testing(z,z1):
    d = np.linspace(-4,4,100)
    n = norm.pdf(d,0,1)
    plt.fill_between(d, n, 0,where =(d<-z1)^(d>z1),alpha=0.3)  
    plt.axvline(x=z, ymin=0, ymax=1, color = 'r', alpha = 0.5)
    plt.xlabel('Z value')
    plt.plot(d, n)
    plt.show()
    
    if z < z1 and z > (-z1):    
        print("\n Conclusion: We are accepting the null hypothesis.\n")
    else:
        print("\n Conclusion: We are rejecting the null hypothesis.\n")
```

#### Hypothesis statement
- Ho: mu = 100 ; Raw corn-starch does not have effect on blood sugar level.
- H1: mu ≠ 100 ; Raw corn-starch does have effect on blood sugar level.
- Critical value of z for two-tailed test at 5% level of significance is ± 1.96

```python
mu = 100
sd = 15
n = 36
x = 108
α = 0.05
z1 = 1.96
```

```python
z = (x-mu)/(sd/math.sqrt(n))
round(z,2)
```

```python
hypothesis_testing(z,z1)
```

Q2. In one state, 52% of the voters are Republicans, and 48% are Democrats. In a second
state, 47% of the voters are Republicans, and 53% are Democrats. Suppose a simple
random sample of 100 voters are surveyed from each state.
What is the probability that the survey will show a greater percentage of Republican
voters in the second state than in the first state?

```python
R1,D1,n1 = 0.52, 0.48, 100
R2,D2,n2 = 0.47, 0.53, 100
mu = R1-R2
x = 0
sd = math.sqrt((R1*D1/n1)+(R2*D2/n2))
z = x-mu/sd
round(z,3)
```

```python
d = np.linspace(-4,4,100)
n = norm.pdf(d,0,1)
p = round(st.norm.cdf(z),2)
plt.fill_between(d, n, 0,where =(d<z),alpha=0.3) 
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
print('The probability that the survey will show a greater percentage of Republican voters in the second state than in the first state',p)
```

Q3. You take the SAT and score 1100. The mean score for the SAT is 1026 and the standard
deviation is 209. How well did you score on the test compared to the average test taker?

```python
mean = 1026
sd = 209
x = 1100
z = (x-mean)/sd
p = round(st.norm.cdf(z),2)
print('Scored',round(z,2),'times more than the average score and is at top',(1-p)*100,'percentile.')
```
