Q 3. The number of times I go to the gym in weekdays, are given below along with its
associated probability:

x = 0, 1, 2, 3, 4, 5

f(x) = 0.09, 0.15, 0.40, 0.25, 0.10, 0.01

Calculate the mean no. of workouts in a week. Also evaluate the variance involved in
it.

```python
x = [0, 1, 2, 3, 4, 5]
p = [0.09, 0.15, 0.40, 0.25, 0.10, 0.01]
import pandas as pd
workout_data = {
    'No_of_times':x,
    'Probability':p
}
df = pd.DataFrame(workout_data)
df['E'] =df['No_of_times']*df['Probability']
Mean = df['E'].sum()
df['V']= (df['No_of_times']-Mean)**2
df['var'] = df['Probability']*df['V']
variance = df['var'].sum()
print('Answer: Mean and variance of number of workouts in a week are', Mean,'and', variance)
df
```

Q.5 A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6
chosen LEDs, what is the probability of having 2 faulty LEDs in my sample?
Calculate the average value of this process. Also evaluate the standard deviation
associated with it.

```python
from scipy.stats import binom 
import matplotlib.pyplot as plt
import math
import numpy as np
n = 6      # no of trials
p = 0.3    # faulty rate
n_values = list(range(n + 1))                     
dist = [binom.pmf(r, n, p) for r in n_values ] 
plt.xlabel('Number of LEDs')
plt.ylabel('Probability')
plt.bar(n_values, dist) 
plt.show()
prob = binom.pmf(2, n, p)
mean, var = binom.stats(n, p)
print('Probability of having 2 faulty LEDs in the sample is',round(prob,4))
print("Average = "+str(np.round(mean, decimals = 4))) 
print('variance = '+ str(np.round(var, decimals = 4)))
print("standard deviation =",round(math.sqrt(var), 4))
print("\nr\tp(r)")
for i in range(n + 1): 
    print(str(n_values[i]) + "\t" + str(np.round(dist[i], decimals = 4))) 
```

The two main governing factors affecting their ability to solve questions correctly are-

Q 7. Customers arrive at a rate of 72 per hour to my shop. What is the probability of k
customers arriving in 4 minutes?
a) 5 customers, b) not more than 3 customers, 
c)more than 3 customers. Give a pictorial representation of the same to validate your
answer.

```python
from scipy.stats import poisson
m = 72/60*4
p1=poisson.pmf(k=5,mu = m)                         # probability of arriving of 5 customers in 4 minutes
p2=sum(poisson.pmf(k = list(range(4)),mu = m))     # probability of arriving of not more than 3 customers in 4 mins 
print('Probability of arriving of 5 customers in 4 minutes is',round(p1,4))
print('probability of arriving of not more than 3 customers in 4 mins is',round(p2,4))
print('probability of arriving of more than 3 customers in 4 mins is',round((1-p2),4))
```

Q 10.Please compute the following:

a) P(Z > 1.26), . P(Z < −0.86), P(Z > −1.37), P(−1.25 < Z < 0.37), . P(Z ≤ −4.6)

b) Find the value z such that P(Z > z) = 0.05

c) Find the value of z such that P(−z < Z < z) = 0.99

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as st
from scipy.stats import norm
d = np.linspace(-4,4,100)
n = norm.pdf(d,0,1)
```

a.1) P(Z > 1.26)

```python
a1 = 1-st.norm.cdf(1.26)
print('Probabilty of Z > 1.26 is',round(a1,3))
plt.fill_between(d, n, 0,where =(d>1.26),alpha=0.5) 
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
```

a.2) P(Z < −0.86)

```python
a2 = st.norm.cdf(-0.86)
print('Probability of Z < −0.86 is',round(a2,3))
plt.fill_between(d, n, 0,where =(d<-0.86),alpha=0.3) 
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
```

a.3) P(−1.25 < Z < 0.37)

```python
a3 = st.norm.cdf(0.37) - st.norm.cdf(-1.25)
print('Probabilty of −1.25 < Z < 0.37 is',round(a3,3))
plt.fill_between(d, n, 0,where =(d<0.37)^(d<-1.25),alpha=0.3)
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
```

a.4) P(Z ≤ −4.6)

```python
a4 = st.norm.cdf(-4.6)
print('Probability of Z ≤ −4.6 is',round(a4,3))
```

b) Find the value z such that P(Z > z) = 0.05

```python
z2 = -round(st.norm.ppf(.05),3)
print('Value of z is', z2)
plt.fill_between(d, n, 0,where =(d>z2),alpha=0.3)  
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
```

c) Find the value of z such that P(−z < Z < z) = 0.99

```python
z3 = st.norm.ppf((1+.95)/2)
print('Value of z is', round(z3,2))
plt.fill_between(d, n, 0,where =(d<z3)^(d<-z3),alpha=0.2)  
plt.xlabel('Z value')
plt.ylabel('Probabilty density')
plt.plot(d, n)
plt.show()
```

Q 12. The shaft in a piston has its diameter normally distributed with a mean of 0.2508 inch
and a standard deviation of 0.0005 inch. The specifications of the shaft are 0.2500 ∓
0.0015 inch. What proportion of shafts are in sync with the specifications? If the
process is centered so that the mean is equal to the target value of 0.2500, what
proportion of shafts conform to the new specifications? What is your conclusion from
this experiment?

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as st
from scipy.stats import norm
d1 = 0.25+0.0015
d2 = 0.25-0.0015
d1,d2
```

```python
m1 = 0.2508
sd1 = 0.0005
s1 = st.norm.cdf((d1-m1)/sd1) - st.norm.cdf((d2-m1)/sd1)
print(round(s1*100,2), 'percents of shafts are in sync with the the specifications.')
d = np.linspace(0.2475,0.2545,100)
n = norm.pdf(d,0.2508,0.0005)
plt.fill_between(d,n,0, where =((d>d1)^(d>d2)),color='blue',alpha=0.15)
plt.xlabel('Diameter of shaft')
plt.plot(d,n)
plt.show()
```
